Este código:
Define un funnel manualmente
Calcula cuántos usuarios únicos están en cada etapa
Imprime los resultados de cada etapa en orden
Sirve para:
Ver pérdidas entre etapas (drop-off)
Analizar el rendimiento del embudo
Evaluar conversiones
Preparar datos para un análisis A/B basado en etapas

```python
print("\n=== Conversión entre etapas ===")

conversion_rates = {}

for i in range(1, len(funnel)):
    prev_step = funnel[i-1]
    curr_step = funnel[i]

    prev_users = funnel_users[prev_step]
    curr_users = funnel_users[curr_step]

    if prev_users > 0:
        rate = curr_users / prev_users
        conversion_rates[curr_step] = rate
        print(f"De {prev_step} → {curr_step}: {rate:.3%}")
    else:
        conversion_rates[curr_step] = None
        print(f"De {prev_step} → {curr_step}: No calculable")
```

Este código:
Calcula la tasa de conversión entre cada etapa del funnel
Imprime la conversión como porcentaje
Guarda las tasas en conversion_rates
Detecta y maneja casos donde no hay usuarios previos
Sirve para:
Analizar caídas entre etapas
Identificar dónde se pierde más tráfico
Evaluar impacto de cambios UX / A/B tests
Reportes de rendimiento del funnel

```python
first_step = funnel[0]
last_step = funnel[-1]

print("\n=== Conversión final del embudo ===")

if funnel_users[first_step] > 0:
    final_conversion = funnel_users[last_step] / funnel_users[first_step]
    print(f"Conversión completa del embudo: {final_conversion:.3%}")
else:
    print("No se puede calcular la conversión final (0 usuarios en primera etapa).")
```

Este código:
Calcula el porcentaje de usuarios que completan todo el funnel
Usa la primera y última etapa definidas en funnel
Imprime el resultado en porcentaje
Maneja el caso de que el funnel pueda estar vacío o mal cargado
Sirve para:
Evaluar la eficacia total del embudo
Comparar conversiones entre grupos A/B
Identificar problemas graves de retención

```python

print("\n=== Pérdidas por etapa ===")

losses = {}

for i in range(1, len(funnel)):
    prev = funnel[i-1]
    curr = funnel[i]
    losses[curr] = funnel_users[prev] - funnel_users[curr]
    print(f"Pérdida de {prev} → {curr}: {losses[curr]} usuarios")
```

Este código:
Calcula cuántos usuarios se pierden entre cada paso del embudo

Imprime los números para identificar cuellos de botella
Sirve para:

Optimizar UX o flujo de conversión
Analizar efectividad del funnel

```python

worst_stage = max(losses, key=losses.get)

print(f"\nPeor etapa del embudo: {worst_stage}")
```

```python
most_popular_event = df['event'].value_counts().idxmax()
print("\nEvento más popular:", most_popular_event)
```

Este código encuentra el evento más común o popular en tu dataset.
Es útil para:
Analizar qué interacción ocurre más frecuentemente.
Validar comportamiento de usuarios.
Preparar análisis de funnels, priorizando eventos clave.

```python

from statsmodels.stats.proportion import proportions_ztest

def compare_groups(event_name, g1, g2):
    users_g1 = df[(df['experiment'] == g1) & (df['event'] == event_name)]['user_id'].nunique()
    users_g2 = df[(df['experiment'] == g2) & (df['event'] == event_name)]['user_id'].nunique()

    total_g1 = df[df['experiment'] == g1]['user_id'].nunique()
    total_g2 = df[df['experiment'] == g2]['user_id'].nunique()

    counts = [users_g1, users_g2]
    nobs = [total_g1, total_g2]

    stat, p_value = proportions_ztest(counts, nobs)

    return {
        "grupo_1": g1,
        "grupo_2": g2,
        "usuarios_g1": users_g1,
        "usuarios_g2": users_g2,
        "proporcion_g1": users_g1 / total_g1,
        "proporcion_g2": users_g2 / total_g2,
        "p_value": p_value
    }
```

Esta función permite comparar la tasa de un evento entre dos grupos experimentales y determinar si la diferencia es estadísticamente significativa usando un proportion Z-test.

```python

result_AA = compare_groups(most_popular_event, 246, 247)

print("\n=== Test A/A para el evento más popular ===")
print(result_AA)
```

Este código:
Compara la proporción de usuarios que realizaron el evento más popular entre dos grupos “idénticos” (246 vs 247).
Evalúa si la diferencia es significativa con un Z-test de proporciones.
Permite validar que la asignación de usuarios a grupos experimentales es correcta antes de realizar un test A/B real.

```python
event_list = df['event'].unique()

AA_results = []

for ev in event_list:
    AA_results.append(compare_groups(ev, 246, 247))

print("\n=== Comparación evento por evento (A/A) ===")
for r in AA_results:
    print(r)
```

Compara proporciones de usuarios por evento entre grupo 248 y grupo 247.
Calcula el valor p para cada evento usando un Z-test de proporciones.
Permite detectar diferencias significativas en el comportamiento de los usuarios entre estas dos variantes de tu experimento.
Un valor p < 0.05 indica que la diferencia entre los grupos es estadísticamente significativa para ese evento.

```python
df['control'] = df['experiment'].apply(lambda x: "control" if x in [246, 247] else "test")

print("\nUsuarios combinados en grupo control:", df[df['control']=="control"]['user_id'].nunique())
print("Usuarios en el grupo 248:", df[df['experiment']==248]['user_id'].nunique())
```

Crea un grupo de control combinado de los grupos 246 y 247.
Clasifica el grupo 248 como grupo test.
Muestra el número de usuarios únicos en control y test.
Esto es útil para análisis A/B donde quieres comparar una variante test contra un control combinado para aumentar la potencia estadística.

```python
def compare_control_combined(event_name):
    users_control = df[(df['control']=="control") & (df['event']==event_name)]['user_id'].nunique()
    users_248 = df[(df['experiment']==248) & (df['event']==event_name)]['user_id'].nunique()

    total_control = df[df['control']=="control"]['user_id'].nunique()
    total_248 = df[df['experiment']==248]['user_id'].nunique()

    counts = [users_248, users_control]
    nobs = [total_248, total_control]

    stat, p_value = proportions_ztest(counts, nobs)

    return {
        "evento": event_name,
        "usuarios_248": users_248,
        "usuarios_control": users_control,
        "proporcion_248": users_248 / total_248,
        "proporcion_control": users_control / total_control,
        "p_value": p_value
    }

results_combined = [compare_control_combined(ev) for ev in event_list]

print("\n=== Comparación 248 vs controles combinados ===")
for r in results_combined:
    print(r)
```

Este código:
Compara grupo test 248 vs control combinado (246+247).
Calcula proporciones de usuarios que realizaron cada evento.
Evalúa si la diferencia es estadísticamente significativa usando un Z-test de proporciones.
